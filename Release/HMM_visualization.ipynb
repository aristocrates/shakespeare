### Train the HMM

```python
# use a fixed random seed so results can be reproduced
seed = 1539525
shakespeare_hmm.random.seed(seed)
shakespeare_hmm.HMM.np.random.seed(seed)
shakespearean_hmm, sonnet = shakespeare_hmm.do_hmm(num_states = 12,
                                                   verbose = False,
                                                   give_hmm = True)
```

```python
# print out the poem
print("Generated Sonnet:\n")
print(sonnet)
print("\nRandom seed: %s" % str(seed))
```

```python
wordclouds = states_to_wordclouds(shakespearean_hmm, shakespeare_hmm.token2index)
```

```python
anim = animate_emission(shakespearean_hmm, shakespeare_hmm.token2index,
                        M=8)
HTML(anim.to_html5_video())
```
